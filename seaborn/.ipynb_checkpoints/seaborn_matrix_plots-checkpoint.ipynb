# Seaborn Matrix Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
# read the tips sample data set
tips = sns.load_dataset("tips")

tips.head()
```

```python
# read the flights sample data set
flights = sns.load_dataset("flights")

flights.head()
```

```python
# return the tips as a correlation (relations) data matrix
tips_correlations = tips.corr()

tips_correlations
```

```python
"""
create a heatmap for the tips_correlations matrix
- annotate the data numbers into the heat cells
- set the cmap (color map) as coolwarm
"""
sns.heatmap(tips_correlations, annot=True, cmap="coolwarm")
```

```python
"""
create a pivot table of the flight with the index as the months, the
columns as the years, and the data as the passengers
- now we can correlate the month and the year to get the data in a cell
"""
flights_pivot = flights.pivot_table(index="month",
                                    columns="year", values="passengers")

flights_pivot
```

```python
"""
create a heatmap for the flights_pivot matrix, you can see as over the years
there were more flights, andyou can also seen that June, July, and
August were the most popular months
- set the line color (border) per cell to white
- set the line width (border width) to 1
"""
sns.heatmap(flights_pivot, linecolor="white", linewidth=1)
```

### Cluster Map
Hierarchical clustering to produce a clustered version of a heatmap.
- Clusters the most similar indexes together

```python
"""
cluster map the indexes with similar passengers together
- years 1949 and 1950 are low passenger months while 1955 and 1956
    are the higher passenger months
"""
sns.clustermap(flights_pivot)
```

```python
"""
cluster map the indexes with similar passengers
together as normalized values from 0 to 1 instead
- janurary, feb, and november are low passenger months while
    june and september are more high passenger months
"""
sns.clustermap(flights_pivot, standard_scale=1)
```

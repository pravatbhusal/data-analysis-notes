# Seaborn Grids
Automate subplots based on the features of the data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the flower iris sample data set
iris = sns.load_dataset("iris")

iris.head()
```

```python
# return the unique values of the iris flower
iris["species"].unique()
```

### PairGrids

```python
# return the paired grids of each numerical column of iris, like pairplot()
grids = sns.PairGrid(iris)

# map a scatter plot onto all the grids
grids.map(plt.scatter)
```

```python
# return the paired grids of each numerical column of iris
grids = sns.PairGrid(iris)

# map a dist plot onto the diagonal plots starting from row = 0, col = 0
grids.map_diag(sns.distplot)

# map a scatter plot onto the upper-right plots that were not plotted so far
grids.map_upper(plt.scatter)

# map a kdeplot onto the lower-left plots that were not plotted so far
grids.map_lower(sns.kdeplot)
```

### FacetGrids

```python
# load the tips sample data set
tips = sns.load_dataset("tips")

tips.head()
```

```python
# create a FacetGrid for the row smoker, column time datas
grids = sns.FacetGrid(data=tips, col="time", row="smoker")

# map the distribution of each smoker and time by total_bill
grids.map(sns.distplot, "total_bill")
```

```python
# create a FacetGrid for the row smoker, column time datas
grids = sns.FacetGrid(data=tips, col="time", row="smoker")

# map the distribution of each smoker and time by total_bill (x) and tip (y)
grids.map(plt.scatter, "total_bill", "tip")
```
